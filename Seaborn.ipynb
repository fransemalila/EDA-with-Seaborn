### Loading Libraries and Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
data = pd.read_csv('Documents/Seaborn/data.csv')
```

### Separate Target from features

```python
data.head()
```

```python
col =data.columns
print (col)
```

```python
y = data.diagnosis
drop_cols = ['Unnamed: 32','id','diagnosis']
x = data.drop(drop_cols, axis=1)
x.head()
```

### Plot Diagnosis Distribution

```python
ax = sns.countplot(y, label="Count")
B, M = y.value_counts()
print('Number of Benign Tumors', B)
print('Number of Malignant Tumors', M)
```

```python
x.describe()
```

### Visualizing Standardized Data 

```python
data = x
data_std = (data - data.mean()) / data.std()
data = pd.concat([y, data_std.iloc[:, 0:10]], axis=1)
data = pd.melt(data, id_vars='diagnosis',var_name='features',value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x='features',y='value',hue='diagnosis', data=data, split=True, inner='quart')
plt.xticks(rotation=45)
```

### Violin Plots and Box Plots

```python
data = pd.concat([y, data_std.iloc[:, 10:20]], axis=1)
data = pd.melt(data, id_vars='diagnosis',var_name='features',value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x='features',y='value',hue='diagnosis', data=data, split=True, inner='quart')
plt.xticks(rotation=45)
```

```python
data = pd.concat([y, data_std.iloc[:, 20:30]], axis=1)
data = pd.melt(data, id_vars='diagnosis',var_name='features',value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x='features',y='value',hue='diagnosis', data=data, split=True, inner='quart')
plt.xticks(rotation=45)
```

```python
sns.boxplot(x='features', y='value', hue='diagnosis',data=data)
plt.xticks(rotation=45);
```

### Using Joint Plots for Feature Comparison

```python
sns.jointplot(x.loc[:,'concavity_worst'],
             x.loc[:, 'concave points_worst'],
             kind= 'reg',
             color='#ce1414');
```

### Observing the distribution of values and their variables
